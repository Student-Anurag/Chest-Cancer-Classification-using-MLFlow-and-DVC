```python
def get_product(x, y):
    return x * y
```

```python
get_product(3, 4)
```

```python
get_product(3, "4")
```

```python
from ensure import ensure_annotations
```

```python
@ensure_annotations
def get_product2(x: int, y: int) -> int:
    return x * y
```

```python
get_product2(x=2, y="5")
```
